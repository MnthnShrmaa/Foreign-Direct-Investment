Creating a function to convert Dollars to Rupees

```python
rates = [45.68,47.69,48.39,45.95,44.93,44.27,45.24,40.26,45.99,47.44,45.56,47.92,54.40,60.50,61.14,65.46,67.07]
year = ['2000-01', '2001-02', '2002-03', '2003-04', '2004-05', '2005-06', '2006-07', '2007-08', '2008-09', '2009-10', '2010-11',
       '2011-12', '2012-13', '2013-14', '2014-15', '2015-16', '2016-17']
sector = ['Sector']
def conv (df, year, rates):
  for i in year:
    df[i] = df[i] * rates[year.index(i)]/10
  return df
```

```python
data_dollar = data.copy()
datars= conv(data, year, rates)
datars.style.set_caption("Amount in Indian Crore Rupees").format(precision =2)
```

```python
pivot = pd.melt(data_dollar, id_vars = sector, value_vars= year, var_name = "Year", value_name = "Investment ($ million)", ignore_index = True)
pivot = round(pivot, 2)
pivot
```

```python
pivotrs= pd.melt(datars, id_vars = sector, value_vars= year, var_name = "Year", value_name= "Investment (Rs Crore)", ignore_index = True)
pivotrs = round(pivotrs, 2)
pivotrs
```

```python
merg = pivotrs.merge(pivot, how = "left")
merg
```

```python
sort = merg.sort_values(['Sector', 'Year'], ignore_index = True)
sort
```

```python
sort = sort[['Sector', 'Investment ($ million)', 'Investment (Rs Crore)']].replace(["CONSTRUCTION DEVELOPMENT: Townships, housing, built-up infrastructure and construction-development projects"
                              ,"SERVICES SECTOR (Fin.,Banking,Insurance,Non Fin/Business,Outsourcing,R&D,Courier,Tech. Testing and Analysis, Other)",
                              "TEA AND COFFEE (PROCESSING & WAREHOUSING COFFEE & RUBBER)"],
                               ["CONSTRUCTION DEVELOPMENT","SERVICES SECTOR",'TEA AND COFFEE']                                     )

```

```python
sectorwise = sort.groupby("Sector").sum()
sectorwise.sort_values(by ="Investment ($ million)", ascending = False)
```

```python
merg.to_csv("FDI clean.csv", index = False)
```
